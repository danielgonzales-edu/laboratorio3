#### 8. Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
datos_Madrid = dataset[dataset['Texto'] == 'Madrid']
datos_Madrid
```

```python
x0 = datos_Madrid['X'].values[0]
y0 = datos_Madrid['Y'].values[0]

dataset_sin_madrid = dataset.drop(datos_Madrid.index[0])
dataset_sin_madrid
```

```python
#utilización de pitágoras y componentes
dataset_copia = dataset_sin_madrid[['Municipio', 'X', 'Y']].copy()
dataset_copia['CX'] = dataset_copia['X']-x0
dataset_copia['CY'] = dataset_copia['Y']-y0
dataset_copia['R'] = np.square((np.power(dataset_copia['CX'], 2)) + (np.power(dataset_copia['CY'],2)))
dataset_copia
```

```python
cercano = dataset_copia[['Municipio', 'R']][dataset_copia['R']==dataset_copia['R'].min()]
cercano
```

```python
lejano = dataset_copia[['Municipio', 'R']][dataset_copia['R']==dataset_copia['R'].max()]
lejano
```

```python
print(f"El municipio más cercano es: {cercano['Municipio'].values[0]}")
print(f"El municipio más lejano es: {lejano['Municipio'].values[0]}")
```
